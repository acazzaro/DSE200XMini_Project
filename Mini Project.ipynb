<h1> A study on the relationship between economic growth and birth rate </h1>

```python
import pandas as pd
```

```python
# Read in the World Development Indicators Database
data = pd.read_csv('./Desktop/WB_dataset/Indicators.csv')
data.head()
```

```python
# Check if there are NaN values
data.isnull().any()
```

```python
# Filter by fertility and year 2013
data_fertility = (data['IndicatorName'].str.contains('Fertility rate, total')) & (data['Year'] == 2013)
fertility = pd.DataFrame(data[data_fertility])
fertility
```

```python
# Filter by GDP per capita and year 2013
data_gdp = (data['IndicatorName'].str.contains('GDP per capita \(current US')) & (data['Year'] == 2013)
gdp = pd.DataFrame(data[data_gdp])
gdp
```

```python
# Merge fertility and gdp dataframes
final = fertility.merge(gdp, on='CountryCode', how='inner')
final.head(50)[33:]
```

```python
# Sort by higher fertility rate
sorted_by_fertility = final[33:].sort_values(by="Value_x", ascending=False)
sorted_by_fertility.head(50)
```

```python
# Sort by higher gdp per capita
sorted_by_gdp = final[33:].sort_values(by="Value_y", ascending=False)
sorted_by_gdp.head(50)
```

```python
#Check for Nan values
final['Value_x'].isnull().values.any()
```

```python
# Check for negative values
final['Value_x'].any() < 0
```

```python
# Check for negative values
final['Value_y'].any() < 0
```
